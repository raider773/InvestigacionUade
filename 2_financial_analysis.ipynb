```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data.csv").drop(columns = ['Unnamed: 0'])
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='estado_id', y='monto_total_adjudicado', data=data)
plt.title('Estados y monto adjudicado')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x='estado_id', y='amount_requested_awarded_difference', data=data)
plt.title('Estados y diferencia solicitado vs adjudicado')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='total_personel', y='monto_total_adjudicado', data=data)
plt.title('monto adjudicado y total personal')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(data['directores'], data['monto_total_adjudicado'], label='directores', s=20)
plt.scatter(data['investigadores'], data['monto_total_adjudicado'], label='investigadores', s=20)
plt.scatter(data['becarios'], data['monto_total_adjudicado'], label='becarios', s=20)
plt.scatter(data['estudiantes'], data['monto_total_adjudicado'], label='estudiantes', s=20)
plt.scatter(data['personales_tecnico_de_apoyo'], data['monto_total_adjudicado'], label='personales_tecnico_de_apoyo', s=20)
plt.scatter(data['co_directores'], data['monto_total_adjudicado'], label='co_directores', s=20)
plt.xlabel('Cantidad de personal')
plt.ylabel('Monto adjudicado')
plt.title('Relación entre cantidad de personal especifico y monto adjudicado')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(data['directores'], data['amount_requested_awarded_difference'], label='directores', s=20)
plt.scatter(data['investigadores'], data['amount_requested_awarded_difference'], label='investigadores', s=20)
plt.scatter(data['becarios'], data['amount_requested_awarded_difference'], label='becarios', s=20)
plt.scatter(data['estudiantes'], data['amount_requested_awarded_difference'], label='estudiantes', s=20)
plt.scatter(data['personales_tecnico_de_apoyo'], data['amount_requested_awarded_difference'], label='personales_tecnico_de_apoyo', s=20)
plt.scatter(data['co_directores'], data['amount_requested_awarded_difference'], label='co_directores', s=20)
plt.xlabel('Cantidad de personal')
plt.ylabel('Diferencia monto')
plt.title('Relación entre cantidad de personal especifico y diferencia solicitado vs adjudicado')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(data['disciplina_id'], data['monto_total_solicitado'],s=50)
plt.xlabel('Disciplina')
plt.ylabel('Monto Solicitado')
plt.title('Relación entre disiplina y monto total solicitado')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(data['disciplina_id'], data['amount_requested_awarded_difference'],s=50)
plt.xlabel('Disciplina')
plt.ylabel('Diferencia solicitado vs adjudicado')
plt.title('Relación entre disiplina y monto total solicitado vs adjudicado')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(data['amount_of_benefactors'], data['monto_total_adjudicado'],s=50)
plt.xlabel('amount_of_benefactors')
plt.ylabel('Monto monto_total_adjudicado')
plt.title('Relación entre can benefactors y monto total adjudicado')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.scatter(data['amount_of_benefactors'], data['amount_requested_awarded_difference'],s=50)
plt.xlabel('amount_of_benefactors')
plt.ylabel('mount_requested_awarded_difference')
plt.title('Relación entre cant benefactors y monto total adjudicado')
plt.legend()
plt.grid(True)
plt.show()
```
