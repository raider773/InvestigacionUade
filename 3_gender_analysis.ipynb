```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data.csv").drop(columns = ['Unnamed: 0'])
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='sexo_director', y='amount_requested_awarded_difference', data=data)
plt.title('Financiamiento por Género del Director')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='sexo_director', hue='estado_id', data=data)
plt.title('Estado por Género del Director')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='sexo_director', y='cantidad_miembros_M', data=data, color='skyblue', alpha=0.8, label='Hombres')
sns.barplot(x='sexo_director', y='cantidad_miembros_F', data=data, color='salmon', alpha=0.6, label='Mujeres')
plt.title('Cantidad de Personal por Género del Director')
plt.legend()
plt.show()
```

```python
#lo de arriba pero por  disiplina estaria muy bueno
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='sexo_director', y='total_personel', data=data)
plt.title('Total de Personal por Género del Director')
plt.show()
```

```python
data["duration"] = pd.to_timedelta(data['duration']).dt.days.astype(int)
```

```python
duration_data = data[["sexo_director","duration"]].groupby("sexo_director").mean()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='sexo_director', y='duration', data=duration_data)
plt.title('Total de duracion por Género del Director')
plt.show()
```

```python
count_data = data.groupby(['disciplina_id', 'sexo_director']).size().reset_index(name='counts')
plt.figure(figsize=(12, 8))
sns.barplot(x='disciplina_id', y='counts', hue='sexo_director', data=count_data)
plt.title('Cantidad de Directores por Disciplina y Género')
plt.xlabel('Disciplina')
plt.ylabel('Cantidad de Directores')
plt.xticks(rotation=45)
plt.legend(title='Género')
plt.tight_layout()
plt.show()
```
